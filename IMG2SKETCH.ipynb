```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

**LOADING ORIGINAL IMAGE**

```python
image=cv2.imread('/content/HULK.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

**CONVERTING IMAGE TO GRAYSCALE**

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(image_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

**INVERTING OF IMAGE**

```python
image_invert = cv2.bitwise_not(image_gray)
plt.figure(figsize=(8,8))
plt.imshow(image_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

**SMOOTHENING OF IMAGE**

```python
img_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

**IMAGE INTO PENCIL SKETCH**

```python
fin = cv2.divide(image_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(fin,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
